# 05. Creating User-Defined Functions

```python
#Defining Functions on 'Price_Label'

def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

```python
#Apply the Function

df['price_range'] = df.apply(price_label, axis=1)
```

# 10. Creating the New Label Variable 'Busiest Period of the Day'


```python
#Creating the New Label Variable 'Busiest Period of the Day';

result_orders = []

for value in df_orders_products_T46 ['order_hour_of_day']:
    if value >=9 and value <=16:
        result_orders.append ('Most Orders')
    elif value <=6:
        result_orders.append ('Fewest Orders')
    else:
        result_orders.append ('Average Orders')
```

```python
df_orders_products_T46 ['Busiest_period_of_day'] = result_orders
```
